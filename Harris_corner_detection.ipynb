```python
import cv2
from google.colab.patches import cv2_imshow
import os
import sys
import numpy as np
img=cv2.imread('/content/0CA41A93-47F3-40CD-9479-8FB4CCFED190 - vinay kumar.jpeg')

def cornerHarris(img):

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    gray = np.float32(gray)
    dst = cv2.cornerHarris(gray, 2, 3, 0.04)

    dst = cv2.dilate(dst, None)

    img[dst > 0.01*dst.max()] = [0, 0, 255]

    cv2_imshow(img)
    cv2.waitKey(0)


#img = cv2.imread(sys.argv[1])
h, w = img.shape[:2]
img = cv2.resize(img, (int(w/10), int(h/10)))


cornerHarris(img)
cv2.imwrite('/content/0CA41A93-47F3-40CD-9479-8FB4CCFED190 - vinay kumar.jpeg', img)
```
